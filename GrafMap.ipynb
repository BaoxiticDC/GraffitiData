```python
import pandas as pd
import os
import cv2
import numpy as np
import time
import pickle
import folium
from folium import plugins
```

```python
# Leer ed Diccionario con la información.

df = pd.read_pickle("dfgrafiti3.pkl")

#diseño del mapa

dfg = df.query('3<n_graf')
data = dfg.groupby(['lat','long'])['n_graf'].sum().reset_index().values.tolist()
heat_data = dfg.groupby(['lat','long'])['n_graf'].sum().reset_index().values.tolist()

m = folium.Map([6.245, -75.537], zoom_start=13)#, tiles = 'stamentoner')

#folium.plugins.HeatMap(heat_data).add_to(m)

folium.plugins.HeatMap(data,name=None, min_opacity=0.5, 
                       max_zoom=18, max_val=8, radius=15, blur=14, 
                       gradient=None, overlay=True, control=True, show=True).add_to(m)
#despliegue del mapa

m
```
